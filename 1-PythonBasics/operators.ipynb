```python
## Arithmetic Operations
a = 10
b = 3  
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
print("Mixed Operations:", (a + b) * (a - b) / b)
print("Operator Precedence:", a + b * a - b / a)
```

```python
print(5/0)  # This will raise a ZeroDivisionError
```

```python
## Comparison Operators
x = 10
y = 20
print("Equal:", x == y)
print("Not Equal:", x != y)
```

```python
str1 = "ravi"
str2 = "Ravi"
print("str1 > str2:", str1 > str2) ## Ascii comparision
print("str1 < str2:", str1 < str2)
print("str1 >= str2:", str1 >= str2)
print("str1 <= str2:", str1 <= str2)
print("str1 == str2:", str1 == str2)
print("str1 != str2:", str1 != str2)
print("str1 is str2:", str1 is str2)
print("str1 is not str2:", str1 is not str2)
print("str1 in str2:", str1 in str2)
print("str1 not in str2:", str1 not in str2)

str2 = "ravi"
print(str1==str2)



```

```python
## Logical Operators - AND, OR, NOT
a = True
b = False
print("a and b:", a and b)
print("a or b:", a or b) 
print("not a:", not a)
print("not b:", not b)
print("a and (not b):", a and (not b))
print("b or (not a):", b or (not a))
print("(a and b) or (not a):", (a and b) or (not a))
print("(a or b) and (not b):", (a or b) and (not b))   
print("(not a) and (not b):", (not a) and (not b))
print("(not a) or (not b):", (not a) or (not b))
print("a or (b and a):", a or (b and a))
print("b and (a or b):", b and (a or b))
print("a and a:", a and a)
print("b or b:", b or b)
print("not (a and b):", not (a and b))
print("not (a or b):", not (a or b))
print("a and (b or a):", a and (b or a))
print("b or (a and b):", b or (a and b))
print("a or (not b):", a or (not b))  
print("b and (not a):", b and (not a))
print("not (a or not b):", not (a or not b))
print("not (b and not a):", not (b and not a))
```
