```python
## Integer Example
age = 25
print("Age is:", age)
print("Type of age:", type(age))
```

```python
## Floting Point Example
height = 5.9
print("Height is:", height)
print("Type of height:", type(height))
```

```python
## String data type example
name = "John Doe"
print("Name is:", name)
print("Type of name:", type(name))
```

```python
## boolean data type example
is_student = True
print("Is student:", is_student)
print("Type of is_student:", type(is_student))

print(type(True))
print(type(False))

is_true = bool(1)
print("is_true:", is_true)
print("Type of is_true:", type(is_true))

a = 10
b = 20
print(type(a==b))
print(type(a!=b))
```

```python
## Common Errors
num = "100"
print("Num as string:", num)
print("Type of num as string:", type(num))
num = int(num)
print("Num as integer:", num)
print("Type of num as integer:", type(num))
# num = int("100.5")  # This will raise a ValueError
```
