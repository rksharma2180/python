```python
## Naming Conventions
## Variable names should be descriptive
## They must always start with a letter or an underscore (_)
## They cannot start with a number
## They can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
## Variable names are case-sensitive (age, Age and AGE are three different variables)
## Avoid using Python reserved keywords as variable names (like if, else, while, for, etc.)
## Use snake_case for variable names (e.g., my_variable_name)
```

```python
##Invalid variable names examples (uncomment to see errors)
#1st_place = "Alice"  # Cannot start with a number
#my-variable = 10    # Hyphens are not allowed
#for = 5            # 'for' is a reserved keyword
_myVariable = 20    # Valid but not recommended (camelCase)
my_variable_name = 30  # Recommended (snake_case)
print("_myVariable:",_myVariable)
print("my_variable_name:",my_variable_name)

```

```python
## Understaning Data Types
## Python is dynamically typed language
## Type of the variable is inferred at runtime
age  = 30            # Integer
height = 5.8        # Float
name = "Alice"     # String
is_student = False  # Boolean
print("Type of age:", type(age))
print("Type of height:", type(height))
print("Type of name:", type(name))
print("Type of is_student:", type(is_student))
# --- IGNORE ---
```

```python
age = 25
print(type(age))  # Output: <class 'int'>
age = "Twenty Five"
print(type(age))  # Output: <class 'str'>
# --- IGNORE ---
```

```python
## Dynamic Typing in Python
## Python allows variables to change types dynamically
var = 10
print("Value:", var, "Type:", type(var))
var = "Now I'm a string"
print("Value:", var, "Type:", type(var))

```
