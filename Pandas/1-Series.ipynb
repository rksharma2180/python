Pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structures.

Series - A series is a one dimentional array like object while data frame is two dimentional, size mutable, and potentially hetrogeneous tabular data structure with labeled axes(row, colum)

```python
import pandas as pd
print(pd.__version__)
```

```python
# Series 
import pandas as pd
ser = pd.Series([1,2,3,4,5])
print(ser)
print(type(ser))
```

```python
# Series from dictionary
series = pd.Series({"a": 1, "b": 2, "c": 3})

print(series)
print(type(series))
```

```python
# Series with custom index
import pandas as pd
data = [10,20,30,40,50]
index = ['a','b','c','d','e ']
s = pd.Series(data, index=index)
print(s)
print(type(s))
```

```python
# Series with Default Index 
v = pd.Series(data=[1,2,3,4,5])
print(v)
print(type(v))
```
