```python
# DataFrames 
import pandas as pd
df = pd.DataFrame({"A":[1,2,3,4,5], "B":[6,7,8,9,10]})
print(df)
print(type(df))
```

```python
data = {
    "Name": ["John", "Anna", "James", "Linda"],
    "Age": [28, 22, 35, 32],
    "City": ["New York", "Paris", "London", "Berlin"]
}
print(data)
print(type(data))
```

```python
# Similar to data frames using numpy
import numpy as np
print(np.array(data)) 
```

```python
# DataFrames 
pd_data = pd.DataFrame(data)
print(pd_data)
print(type(pd_data))
```

```python
# Data Frame using list of dictories
list_data = [
    {"Name": "John", "Age": 30, "City": "New York"},
    {"Name": "Anna", "Age": 22, "City": "London"},
    {"Name": "Mike", "Age": 32, "City": "San Francisco"}
]
df = pd.DataFrame(list_data)
print(df)
print(df['Name'])
print(type(df['Name']))
print(df.iloc[0]['Age']) # 2nd row 1st column is basically label based indexing
```

```python
print(df.loc[0])
print(df.iloc[0])
print(df.iloc[0,0])
print(df.iloc[0,1])

# print(df.iloc[0][1]) # KeyError
```
