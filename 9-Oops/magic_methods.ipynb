Magic Methods
Magic methods in python, also known as Dunder methods (dobule underscore method), are special method that start and end with double underscore.
These methods enable us to define the behavior of objects for built in operations such as arithmeric operation, comparison and more. 

```python
'''
Magic methods are predefined methods in python that you can override to chang the behavior of our object. some common magic methods
- __init__: Initialize the new instance of a class
- __str__: returns string represenatation of a class
- __repr__: Returns an official string representation of a class
- __len__: returns length of an object
- __getitem__: gets an item from a container
- __setitem__: Sets an item in a container
'''
```

```python
class Person:
    pass

person = Person()
dir(person) # list of megic methods and attributes
```

```python
class Person:
    pass

person = Person() 
print(person) # <__main__.Person object at 0x000001FDF5A3C830> default implementation of to string
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):  # Overriding to string method
        return f'{self.name}  {self.age}'

person = Person('Ravi', 25)
print(person)    
```
