```python
## How to import a module/package

import math
print(math.sqrt(16))

## How to import a specific function from a module/package

from math import sqrt
print(sqrt(16))
```

```python
from math import factorial

print(factorial(5))
```

```python
## numpy is package provides helper methods for array operations
## to create or perform array
import numpy as np

print(np.array([1,2,3,4,5]))
```

```python
from math import * ## Import all the function from math module

print(sqrt(25))
print(pi)
```

```python
## Importing custom package
from package.math import addition

print(addition(5,6))
```

```python
from package import math

print(math.addition(18,22))
print(math.subtraction(14,7))
```

```python
from package.math import *

print(addition(5,5))
print(subtraction(7,2))
print(multiplication(4,5))
print(division(12,4))
```
