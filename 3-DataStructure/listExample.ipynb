```python
## Create a todo list task 
todo_list = []
todo_list.append("Buy groceries")
todo_list.append("Clean the house")
todo_list.append("Pay bills")

print("Todo List:", todo_list)

todo_list.append("Setup a meeting with team members")
todo_list.append("Finish the project report")   
print("Updated Todo List:", todo_list)

todo_list.remove("Clean the house")  ## Remove task from todo list
print("After removing a task, Todo List:", todo_list)

if "Pay bills" in todo_list:
    print("Task 'Pay bills' is in the todo list.")
else:
    print("Task 'Pay bills' is not in the todo list.")

```

```python
grades = [85, 90, 78, 92, 88]
grades.append(95)  ## Add a new grade
average_grade = sum(grades) / len(grades)
print("Average Grade:", average_grade)

print(max(grades))  ## Highest grade
print(min(grades))  ## Lowest grade
grades.sort()  ## Sort grades in ascending order
print("Sorted Grades:", grades)
```

```python
feedback = ["Great product!", "Could be improved.", "Excellent service.", "Not satisfied."]
for comment in feedback:
    print("Customer Feedback:", comment)
positive_feedback = [comment for comment in feedback if "Great" in comment or "Excellent" in comment]
print("Positive Feedback:", positive_feedback)

positive_feedback_count = sum(1 for comment in feedback if "Great" in comment or "Excellent" in comment)
print("Number of Positive Feedbacks:", positive_feedback_count)
```
