```python
## Sets
## Sets are unordered collections of unique elements
## Sets are mutable, meaning you can add or remove elements
## Sets do not allow duplicate elements 
## sets are defined using curly braces {} or the set() function
## sets are useful for membership testing and eliminating duplicate entries from a list
```

```python
## create a set
my_set = {1, 2, 3, 4, 5}
print("Initial Set:", my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print("Empty Set:", my_empty_set)
print(type(my_empty_set))   
```

```python
filled_set = set([1, 2, 3, 4, 5])
print("Filled Set using set():", filled_set)    
```

```python
unique_set = {1, 2, 2, 3, 4, 4, 5}
print("Set with duplicates removed:", unique_set)
```

```python
## Basic Set Operations
my_set.add(6)
print("After adding 6:", my_set)
my_set.remove(4)
print("After removing 4:", my_set)
my_set.add(1)
print("After adding duplicate 1:", my_set)

```

```python
my_set.discard(10)  # Does not raise an error if element not found
print("After discarding 10 (not present):", my_set) 
```

```python
my_set.pop()   # Raises KeyError if element not found
print("After popping 7 (not present):", my_set)
```

```python
## Clear all the elements from the set
my_set.clear()
print("After clearing the set:", my_set)
```

```python
## Set membership testing
my_set = {1, 2, 3, 4, 5}
print("Is 3 in the set?", 3 in my_set)
print("Is 10 in the set?", 10 in my_set)    
```

```python
## Mathametical Set Operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}


###
# Union
set_union = set_a.union(set_b)
print("Union of A and B:", set_union)
```

```python
intersection = set_a.intersection(set_b) ## It give common set elements 
print("Intersection of A and B:", intersection)
```

```python
## Intersection update 
set_a.intersection_update(set_b)
print("Set A after intersection update with B:", set_a)
difference = set_a.difference(set_b)
print("Difference of A and B (A - B):", difference)
symmetric_difference = set_a.symmetric_difference(set_b)
print("Symmetric Difference of A and B:", symmetric_difference)
## Symmetric difference update
set_a.symmetric_difference_update(set_b)
print("Set A after symmetric difference update with B:", set_a)
```

```python
## Set Methods Summary
# add(), remove(), discard(), pop(), clear(), union(), intersection(), intersection_update(), difference(), symmetric_difference(), symmetric_difference_update()
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.issubset(set2))  # Check if set1 is a subset of set2
print(set1.issuperset(set2))  # Check if set1 is a superset of set2
```

```python
ls = [1,2,2,3,4,4,5,5]
unique_ls = set(ls)
print("Unique elements from list:", unique_ls)
```

```python
## Counting unique words in a sentence
sentence = "This is a sample sentence with several words this is a test"
print(f"total words in the sentence: {len(sentence.split())}")
words = sentence.lower().split()
unique_words = set(words)
print("Unique words in the sentence:", unique_words)
print("Number of unique words:", len(unique_words))
```
