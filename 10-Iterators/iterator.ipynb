```python
# Iterators are advance python concept allow for efficient looping and memory management. 
# Iterators provide a way to access elements of a collection sequentially without exposing the underlying 
# structure. Iterator uses lazy loading technique. 
```

```python
list = [2,3,4,5,6]
itr = iter(list) 
print(itr)
```

```python
# In order to iterate on a list using iterator we need to use next method.
next(itr)
```

```python
try:
    print(next(itr))
except StopIteration as si:
    print('No more elements')
```
