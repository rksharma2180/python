```python
## Temprature Conversion Function  
def convert_temprature(temprature, unit):
    if unit == "C":
        return (temprature * 9/5) + 32
    elif unit == "F":
        return (temprature - 32) * 5/9
    else:
        return "Invalid unit"

print(convert_temprature(25, "C"))  # Output: 212.0
print(convert_temprature(77, "F"))  # Output: 100.0
```

```python
## Password Strength Checker Function
def check_password_strength(password):
    """ This function checks the strength of a password based on its length. """
    if len(password) < 8:
        return False
    elif not any(char.isdigit() for char in password) or not any(char.isalpha() for char in password):
        return False
    else:
        return True

print(check_password_strength("ABCDEFGHT1224IJSK"))  # Output: Weak
```

```python
## Calculate cost of items in the cart
def calculate_cart_total(cart):
    total = 0
    for item in cart:
        total += item["price"] * item["quantity"]
    return total

cart =[
    {"name": "Apple", "price": 10, "quantity": 3},
    {"name": "Banana", "price": 15, "quantity":5},
    {"name": "Orange", "price": 20, "quantity":2}
]

print(f"Total cost of items in the cart: {calculate_cart_total(cart)}")
```
