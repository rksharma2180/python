```python
## A map function applies a given function to all items in an input list 
## (or any other iterable) and returns a map object. This is particularly useful 
## for data transforming in a list comprehansively  


## Lazy loading. Make an iterator that computes the function using arguments 
## from each of the iterables. Stops when the shortest iterable is exhausted.
numbers = [1,2,3,4,5,6]
list(map(lambda x:x**2, numbers))  ## Lambda function with map
```

```python
## map() takes two parameters 1. function 2nd Iterable or Collection
def square(x):
    return x**2

print(square(8))
print(map(square,numbers))
list(map(square,numbers))
```

```python
## Map with multiple iterables
numbers1 = [3,4,5,6,7,8]
numbers2 = [1,2,9,8,7,6]

added_numbers = list(map(lambda x,y:x+y, numbers1, numbers2))

print(added_numbers)
```

```python
## map() to convert a list of String to int
str_numbers = ['1','2','3','4','5']
int_numbers = list(map(int, str_numbers))
print(int_numbers)
```

```python
## inbuilt function in 
words = ['apple', 'banana', 'cherry']
upper_words = list(map(str.upper, words))
print(upper_words)
```

```python
## map function is like stream.map in java
def get_name(person):
    return person.get('name')

people = [
    {"name":"John", "age":36},
    {"name":"Sara", "age":28},
    {"name":"Mike", "age":32}
]
print(get_name(people[0]))

list(map(get_name, people))   
```
