```python
## Abstract Base class

from abc import ABC, abstractmethod ## A decorator indicating abstract methods.

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Bark"

dog = Dog()
print(dog.sound())
```

```python
## Abstract Properties
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @property
    @abstractmethod
    def max_speed(self):
        """Abstract property - must be implemented"""
        pass
    
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def __init__(self):
        self._max_speed = 200
    
    @property
    def max_speed(self):
        return self._max_speed
    
    def start_engine(self):
        return "Car engine started"

class Motorcycle(Vehicle):
    def __init__(self):
        self._max_speed = 180
    
    @property
    def max_speed(self):
        return self._max_speed
    
    def start_engine(self):
        return "Motorcycle engine started"

car = Car()
print(car.max_speed)      # 200
print(car.start_engine()) # Car engine started

bike = Motorcycle()
print(bike.max_speed)      # 180
print(bike.start_engine()) # Motorcycle engine started
```
