Encapsulation is a concept of  Oops which wraps data (Variabls) and Behavior (Methods) into single unit. It restrict direct access to components of class to protect data and behavior 

```python
### Encapsulation with getter and setter methods
### Public, Protected, private variables or access modifiers

class Person:
    def __init__(self, name, age):
        self.name = name # Public Variable
        self.age = age # Public Variable

person = Person('Ravi', '25')
print(f'Name: {person.name} Age: {person.age}')
dir(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name # Private Variable
        self.__age = age # Private Variable
        self.gender = gender

person = Person('Ravi', '25', 'Male')
print(f'Gender: {person.gender}')
dir(person)
# print(f'Name: {person.name} Age: {person.age}') # AttributeError: 'Person' object has no attribute 'name'
```

```python
def get_person_details(person):
    return person.__name # AttributeError: 'Person' object has no attribute '__name'

person1 = Person("John", 30, 'Male')
print(get_person_details(person1))
```

```python
class Student:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def __str__(self): # Accessing private data using public method
        return f"Student(name={self.__name}, age={self.__age})"

student = Student("John", 20)
print(student)
```

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self._account_number = account_number  # Protected (convention)
        self._balance = balance                # Protected (convention)
    
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            return True
        return False
    
    def get_balance(self):
        return self._balance

account = BankAccount("123456", 1000)
print(account.get_balance())  # 1000 - proper access

# Can still access directly (but shouldn't - it's a convention)
print(account._balance)  # 1000 - works but violates convention
account._balance = -500  # Bad practice but Python allows it
```

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    def __str__(self):
        return f"Person(name={self._name}, age={self._age})"
    
class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary
    
    def __str__(self):
        return f"Employee(name={self._name}, age={self._age}, salary={self.salary})"

employee = Employee("John", 30, 5000)
print(employee)
print(employee._name) # Accessing protected variable
```

```python
# Encapsulation with Getter and Setters
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    
    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name
    
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        self.__age = age


    def __str__(self): # toString() in Java
        return f"Person(name={self.__name}, age={self.__age})"

person = Person("John", 30)
print(person)
person.set_name("Jane")
person.set_age(25)
print(person)
```
