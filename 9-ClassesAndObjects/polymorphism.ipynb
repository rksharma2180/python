##### Polymorphism is a core concept of OOP that allows objects to be treated as instances of their parent class. It provides a way to perform a single action in different ways/forms. Polymorphism is the ability of an object to take on many forms. It is the ability of a function to perform different actions based on the object it is acting upon. 

##### It is achieved by method overloading and interfaces. 

```python
## Method overriding
## Allows a child class to provide a specific implementation of a method
## that is already defined in its parent class.     
# Base class
class Animal:
    def speak(self):
        print("Animal speaking")

# Derived class
class Dog(Animal):
    def speak(self):
        print("Dog barking")

dog = Dog()
dog.speak()  # Output: Dog barking

# Derived class

class Cat(Animal):
    def speak(self):
        print("Cat meowing")

cat = Cat()
cat.speak()  # Output: Cat meowing

def animal_sound(animal): ## Module level function
    animal.speak()

animal_sound(dog)  # Output: Dog barking
animal_sound(cat)  # Output: Cat meowing
```

```python
# Polymorphism with functions and methods 
class Shape:
    def area(self):
        pass
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self): ## Class Level method
        return 3.14 * self.radius * self.radius
    
class Square(Shape):
    def __init__(self, side):
        self.side = side
    
    def area(self): ## Class Level method
        return self.side * self.side

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self): ## Class Level method
        return 0.5 * self.base * self.height

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self): ## Class Level method
        return self.length * self.width
    
def calculate_area(shape): ## Module level function
    return shape.area()
    
circle = Circle(5)
square = Square(4)
    
print(calculate_area(circle))
print(calculate_area(square))
```
