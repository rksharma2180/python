Inheritance is a fundamental concept in object oriented programming that allows a class to inherit attributes and methods from another class.  

```python
## Inheritance
# Parent Class
class Shape:
    def __init__(self, type):
        print('Parent class initializer called')
        self.type = type

    def draw_shape(self):
        print(f'{self.type} class starting to drawing shape')    

# Child Class
class Rectangle(Shape):## extending parent class
    def __init__(self, length, width):
        super().__init__('Rectangle') ## Calling parent class initializer
        self.length = length
        self.width = width

    def calculate_area(self):
        print(f'Area of {self.type} = {self.length * self.width}')

rec = Rectangle(5, 20)
rec.calculate_area()
rec.draw_shape()

## can parent class call child method with or without reference of child
```

```python
## Multiple inheritance - When a class inherits more than one class 
## Base Class 1
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print('Sub classes must implement this method')

## Base Class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner
    
## Derived class
class Dog(Animal,Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self,owner)

    def speak(self):
        print(f'{self.name} say woof')

dog = Dog('Locy', 'Ravi')
dog.speak()
```
