Object-oriented Programming is programming paradigm that uses "Object" to design applications and programs. Oops allows modelling real world scenarios using classes and objects. 

```python
### A class is a blue print for creating objects. Attribute and methods
```

```python
## Class definition
class Car:
    pass

audi = Car()
print(audi)

bmw = Car()
print(bmw)

## Attributes not recommended way cause it is associated with object
audi.windows = 4 
print(audi.windows)
#print(bmw.windows) # AttributeError: 'Car' object has no attribute 'windows'

## The dir() function in Python is a built-in function that returns a list 
# of valid attributes and methods of an object. It's extremely useful for 
# exploring objects, classes, and modules.

# Without arguments - returns names in current scope
print(len(dir()))

print("=============")

# With an object - returns attributes and methods of that object
dir(audi) ## dir used for debugging 
```

```python
## Instance Variable and Instance methods
class Animal:
    ## Constructor
    def __init__(self, name, age):
        self.name = name ## self is used to access instance variable
        self.age = age
        print('Constructor call')
    
dog = Animal('my dog', 10)
print(dog.name)
print(dog.age)
```

```python
dog2 = Animal('Lucy', 4)
print(dog2.name)
print(dog2.age)
```

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        print(f"{self.name} is Barking")
    
dog = Dog('Lucy', 4)
dog.bark() ## Self is passed automatically
```
