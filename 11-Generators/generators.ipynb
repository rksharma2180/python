```python
# Generater's are simple way of creating iterator. It uses yeild keyword to produce a series of value
# lazily which means they generate value on the fly and  do not store them into memory.

def square(n):
    for i in range(n):
        yield i ** 2 # Save the value and return

square(3)
```

```python
for i in square(4):
    print(i)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()

#print(next(gen))
#print(next(gen))
#print(next(gen))
```

```python
for val in gen:
    print(val)
```

```python
### Practical: Reading large file 

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

for line in read_large_file('large_file.txt'):
    print(line)
```
