```python
## OS module
import os
cwd = os.getcwd()
print(f"current working directory: {cwd}")
```

```python
## New Directory 
new_dir= "package"
os.mkdir(new_dir)   
```

```python
## List all the files and directory in the directory

items = os.listdir(".") ## List all the files and directory in the current directory
print(items)
```

```python
## Joining paths
dir = "folder"
file = "file.txt"
path = os.path.join(os.getcwd(),dir, file)
print(path)
```

```python
path = "sample.txt"
if os.path.exists(path):
    print("File exists")
else:
    print("File does not exist")
```

```python
## Checking if a path is file or directory 

os.path.isfile(path)
```

```python
## Getting the absolute path

relative_path = "data.txt"
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
