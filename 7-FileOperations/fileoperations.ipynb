```python
## Read a whole file

with open("./example.csv", "r") as file:
    print(file.read()) ## Read entire content of file

## Read a line by line

with open("./example.csv", "r") as file:
    for line in file:
        print(line)
```

```python
with open("./sample.txt", "w") as file: ##write function overwrites the file
    file.write("Hello World")
    file.write("\n This is a new line")
    file.writelines(["\n First Line", "\n 2nd Line", "\n 3rd Line"])
```

```python
with open("./sample.txt", "a") as file:
    file.write("\nThis is a new line")
    
```

```python
## Binary files
with open("example.bin", "rb") as f:
    content = f.read()
    print(content)
```

```python
with open("example.bin", "ab")  as file:
    file.write("\nThis is a new line".encode())
```

```python
## Copy content of a file to another file 
with open("./sample.txt", 'r') as file:
    content = file.read()

with open('./destination.txt', 'w') as file:
    file.write(content)    
```

```python
## Read the file and count lines, words and characters

with open('data.txt', 'r') as file:
    lines = file.readlines()
    words = sum(len(line.split()) for line in lines)
    characters = sum(len(line) for line in lines)

print(f"Lines: {len(lines)}")
print(f"Words: {words}")
print(f"Characters: {characters}")
```

```python
## Writing and reading a file
with open('data.txt', 'w+') as file:
    file.write('Hello world!\n')
    file.write('This is a new line')
    file.seek(0) ## Move the cursor to the start of the file cause cursor
    # is at the end of the file after writing
    print(file.read())
```
