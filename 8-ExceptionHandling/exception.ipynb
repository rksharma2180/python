Exception Handling
Exception handling is a process in which we handle the error in our code.
Exceptions are events that occur during execution of a program that disrupt normal flow of the program.

Common exception include:
ZeroDivisionError
NameError
IndexError
KeyError
ValueError
FileNotFoundError



```python
a=b
print(a) #NameError: name 'b' is not defined
```

```python
## Exception try, except block
try:
    a=b
    print(a)
except NameError as ne:
    print(f'The variable b is not assigned: {ne}' )
```

```python
try:
    print(5/0)
except ZeroDivisionError as ze:  
    print(f"{ze}")
```

```python
arr = []
try:
    print(arr[0])
except IndexError as ie:
    print(ie)
```

```python
dict = {"name":"ravi", "age":"35"}
try:
    print(dict["education"])
except KeyError as ke:
    print(ke)
```

```python
#file = None  # Initialize to None first
try:
    file=open('java.txt','r')
    file.read()
except FileNotFoundError as ff:
    print(ff)
except Exception as e:
    print(e)
finally:
    # if file is not None and not file.closed:
    # locals() Return a dictionary containing the current scope's local variables.
    print(locals())
    #  if file is not None and not file.closed:
    if file in locals() and not file.closed():
        file.close()
    print("file closed")
```

```python
# valueErrorException code example
# Example 1: Converting non-numeric string to int
try:
    number = int("hello") ## Converting non numeric string to int
except ValueError as e:
    print(f"Error: {e}")
```

```python
# Example 2: Converting invalid string to float
try:
    price = float("12.35")
except ValueError as e:
    print(f"Error: {e}")
```

```python
# Too many/few values to unpack
try:
    a,b = [1,2,3]
except ValueError as e:
    print(f"Error: {e}")
```

```python
try:
    numbers = [1,2,3,4]
    numbers.remove(10)
except ValueError as e:
    print(f"Error: {e}")
```

```python
try:
    fruits = ['apple', 'banana', 'orange']
    position = fruits.index('grape')
except ValueError as e:
    print(f"Error {e}")
```

```python
import math
try:
    result = math.sqrt(-1)
except ValueError as e:
    print(f"Error: {e}")
```

```python
try:
    result = math.log(-5)
except ValueError as e:
    print(f"Error: {e}")
```

```python
from datetime import datetime
try:
    date = datetime.strptime("2024-13-01", "%Y-%m-%d")
except ValueError as e:
    print(f"Error: {e}")
```

```python
try:
    2/0
    a=b
except Exception as e: ## Exception is parent class for all exceptions
    print(e)
```

```python
try:
    num = int(input("Enter a number: "))
    result = num/0
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
## Try and except else block
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
else:
    print(f"The square of {num} is {num ** 2}")
```

```python
## Try except finally
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
except Exception as e:
    print(f"An error occurred: {e}")
else:
    print(f"Result: {result}")
finally:
    print("This will always execute.")
```
