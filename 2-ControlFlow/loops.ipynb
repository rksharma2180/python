```python
## Types of Loops in python
# 1. For Loop
# 2. While Loop
# 3. Nested Loop

for i in range(5): ## Range is function that generates a sequence of numbers
    print("Iteration:", i) 
```

```python
for j in range(2, 10): ## start, end
    print("Value:", j)
```

```python
for k in range(1, 20, 3): ## start, end, step
    print("Step Value:", k) 
```

```python
for k in range(10, 0, -1): ## start, end, step
    print("Reverse Step Value:", k) 
```

```python
string = "Hello"
for char in string:
    print("Character:", char)
```

```python
## While Loop

count = 0
while count < 5:
    print("Count is:", count)
    count += 1  ## Incrementing the count to avoid infinite loop
```

```python
## Loop control statements
# break, continue, pass
for num in range(10):
    if num == 5:
        break  ## Exits the loop when num is 5
    print("Number before break:", num)
```

```python
## Continue example
for num in range(10):
    if num % 2 == 0:
        continue  ## Skips the even numbers
    print("Odd Number:", num)
    
```

```python
## Pass Example
for num in range(5):
    if num == 3:
        pass  ## Placeholder, does nothing its a null operation
    print("Number with pass:", num)
```

```python
## Nested Loop Example
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```

```python
## Some of 10 natural numbers
i = 0
sum = 0
while i < 11:
    sum += i
    i += 1
print("Sum of natural numbers:", sum)
```

```python
## For else example
for n in range(5):
    print("Number in for-else:", n)
else: ## Executes when the loop is not terminated by break statement 
    print("For loop completed without break")
```

```python
for num in range(1,101):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print("Prime number:", num)
                
```
